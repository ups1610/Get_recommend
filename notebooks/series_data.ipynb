```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import numpy as np
```

Extracting tv_series data from imdb

```python
dataframe_list = []
for num in range(1,9902,50):
    if(num==1):
        link = "https://www.imdb.com/search/title/?title_type=tv_series&ref_=adv_prv"
    else:    
        link = "https://www.imdb.com/search/title/?title_type=tv_series&start="+str(num)+"&ref_=adv_nxt"   
    response = requests.get(link)
    soup = BeautifulSoup(response.content,'html.parser')
    
    series_title = []
    year = []
    rating = []
    genre = []
    actors = []
    votes =[]
    duration = []
    
    series_data = soup.find_all('div',attrs={'class':'lister-item mode-advanced'})
    for data in series_data:
        if data.h3.a:      
            name = data.h3.a.text
        else:
            name = np.NaN    
        if data.h3.find('span',class_='lister-item-year text-muted unbold'):
            s_year = data.h3.find('span',class_='lister-item-year text-muted unbold').text
        else:
            s_year = np.NaN    
        if  data.p.find('span',class_='runtime'):    
            s_runtime = data.p.find('span',class_='runtime').text.replace(' min','')
        else:
            s_runtime = np.NaN 
        if data.p.find('span',class_='genre'):
            s_genre = data.p.find('span',class_='genre').text
        else:
            s_genre = 'unknown'
        if data.find('div',class_='inline-block ratings-imdb-rating'):
           s_rating = data.find('div',class_='inline-block ratings-imdb-rating').text.replace('\n','')
        else:
           s_rating = np.NAN
        if data.find('p',class_=''):
            stars = data.find('p',class_='').text.replace('Stars:','').replace('\n','').strip(' ')
        else:
            stars = 'unknown'   
        if data.find('p',class_='sort-num_votes-visible'):
            vote = data.find('p',class_='sort-num_votes-visible').text.replace('\n','').replace('Votes:','').strip(' ')
        else:
            vote = np.NaN  
        series_title.append(name)
        year.append(s_year)
        actors.append(stars)
        rating.append(s_rating)
        genre.append(s_genre)
    frame_data = {'title':series_title,'year':year,'cast':actors,'rating':rating,'genre':genre}
    df_frame = pd.DataFrame(frame_data)
    df_frame['year'] = df_frame['year'].str.extract(r'(\d{4})')
    df_frame['genre'] = df_frame['genre'].str.replace('\n','')
    dataframe_list.append(df_frame)
```

```python
for df in dataframe_list:
    df_series = pd.concat([df_series,df],ignore_index=True)
```
